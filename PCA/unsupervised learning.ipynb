```python
#replace missing values by median
vehicles_pd=vehicles_pd.replace('inf',np.nan)
autofiller=lambda x:x.fillna(x.median())
vehicles_pd_append=vehicles_pd.drop(['class'],axis=1).apply(autofiller,axis=1)
```

Train Test split of data

```python
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
X=vehicles_pd_append
y=vehicles_pd['class']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
Xscaled_train=x_train.apply(zscore)
Xscaled_test=x_test.apply(zscore)
Xscaled_train.head()
```

PCA

```python
#Finding the cov matrix
cov_mat=np.cov(Xscaled_train,rowvar=False)
cov_mat.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=18)
pca.fit_transform(Xscaled_train)
```

```python
pca.explained_variance_
variance_per=[(i/sum(pca.explained_variance_))*100 for i in pca.explained_variance_]
variance_per=np.cumsum(variance_per)
variance_per
```

```python
sns.barplot(x=[i for i in range(1,len(pca.explained_variance_)+1)],y=pca.explained_variance_)
```

```python
sns.lineplot(x=[i for i in range(1,len(variance_per)+1)],y=variance_per)
```

```python
cumsum=np.cumsum(pca.explained_variance_)
cumsum
```

```python
sns.lineplot(x=[i for i in range(1,len(pca.explained_variance_)+1)],y=cumsum)
```

```python
#from the above graph we can see that 12 principal components can explain 95% of variance in the given data so let us take 12 components for our analysis
pca=PCA(n_components=12)
Xscaled_10_train=pca.fit_transform(Xscaled_train)
Xscaled_10_test=pca.fit_transform(Xscaled_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
svc=SVC()
grid_params={
    'C':[0.1,0.5,1.0],
    'gamma':[0.1,0.2,0.5,1.0],
    'kernel':['rbf','linear','poly']
}
grid=GridSearchCV(svc,grid_params)
```

```python
grid.fit(Xscaled_train,y_train)
```

```python
y_pred=grid.predict(Xscaled_test)
```

```python

scores=[]
cv_scores=[]
cv_avg=[]
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print('accuracy score with raw data ',accuracy_score(y_pred,y_test))
scores.append(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
grid.fit(Xscaled_10_train,y_train)
```

```python
y_pred=grid.predict(Xscaled_10_test)
```

```python
print('accuracy score with PCA ',accuracy_score(y_pred,y_test))
scores.append(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
seed=7
kfold=KFold(n_splits=10,random_state=seed)
scores_crossval=cross_val_score(grid,Xscaled_train,y_train,cv=kfold)
cv_score=scores_crossval.mean()
cv_avg.append(cv_score)
std=scores_crossval.std()
cv_scores.append([cv_score+2*std,cv_score-2*std])
```

```python
scores_crossval=cross_val_score(grid,Xscaled_10_train,y_train,cv=kfold)
cv_score=scores_crossval.mean()
std=scores_crossval.std()
cv_avg.append(cv_score)
cv_scores.append([cv_score+2*std,cv_score-2*std])
```

```python
result_dict={'name':['Raw data','PCA'],
            'scores':scores,
            'estimated Cross Validation scores range':cv_scores,
            'average Cross validation scores':cv_avg}
resultdf=pd.DataFrame(result_dict)
resultdf
```
